```python
#LABELING
import openpyxl as op
from openpyxl.styles import PatternFill as pf
import os
from textblob import TextBlob
from datetime import datetime as dt

netral = 'BCBCBC'
positif = '00FF00'
negatif = 'FF0000'
kolom = []
temp = []
nb = 'labeling'
lokasi = os.getcwd()
dr = '\\sc'
filenya = lokasi+dr+'\\data.xlsx'
head = 'full_text'
ket = "labeling"
output = lokasi+dr+'\\output_'+nb+'.xlsx'
now = dt.now()
tgl = f"\nTanggal dan waktu: {now}\n\n"
wb = op.load_workbook(filenya)
ws = wb.active
max = ws.max_row - 1
a = 0

for aa in ws:
    a += 1
    if a == 1:
        for bb in aa:
            u = bb.value
            p = bb.coordinate
            if u == head:
                o = 1
                for _ in range(max):
                    o += 1
                    kolom.append(f"{p[:1]}{o}")

for k in kolom:
    x = str(ws[k].value)
    labeling = TextBlob(x)
    if labeling.sentiment.polarity > 0.0:
        l = "Positif"
    elif labeling.sentiment.polarity == 0.0:
        l = "Netral"
    elif labeling.sentiment.polarity < 0.0:
        l = "Negatif"
    temp.append({"opini": [x], "label": [l]})

nt = 0
ng = 0
ps = 0

for ww in temp:
    if ww['label'][0] == 'Netral':
        nt += 1
    elif ww['label'][0] == 'Negatif':
        ng += 1
    elif ww['label'][0] == 'Positif':
        ps += 1
print(f"{tgl}Hasil {ket}:\n- Jumlah Netral  : {nt}\n- Jumlah Negatif : {ng}\n- Jumlah Positif : {ps}\nDari total {max} baris\n")

def coloring(ww):
    for dt in temp:
        ww += 1
        opini = dt["opini"][0]
        label = dt["label"][0]
        if label == 'Netral':
            n = pf(patternType='solid', fgColor=netral)
        elif label == 'Negatif':
            n = pf(patternType='solid', fgColor=negatif)
        elif label == 'Positif':
            n = pf(patternType='solid', fgColor=positif)
        sheet.append([opini, label])
        sheet[f"B{ww}"].fill = n

newWB = op.Workbook()
sheet = newWB.active
hd = ["opini", "label"]
sheet.append(hd)
wr = 1
if os.path.exists(output):
    while True:
        if not os.path.exists(output):
            coloring(wr)
            newWB.save(output)
            size = os.stat(output)
            if size.st_size <= (5*1024):
                os.remove(output)
                wr = 1
                coloring(wr)
                newWB.save(output)
                continue
            else:
                print(f"Silahkan cek file output {ket} pada {output}\n")
                break
        os.remove(output)
else:
    coloring(wr)
    newWB.save(output)
    size = os.stat(output)
    if size.st_size <= (5*1024):
        while True:
            os.remove(output)
            wr = 1
            coloring(wr)
            newWB.save(output)
            print(f"Silahkan cek file output {ket} pada {output}\n")
            break
    else:
        print(f"Silahkan cek file output {ket} pada {output}\n")
```

```python
#NORMALISASI
import re

kolom = []
temp = []
teks = ""
nb = "normalisasi"
dr = "\\sc"
lokasi = os.getcwd()
f = "output_stemming.xlsx"
now = dt.now()
tgl = f"\nTanggal dan waktu: {now}\n"
ket = "normalisasi"
filenya = lokasi+dr+"\\"+f
head = "opini"
m = "output_"+nb+".xlsx"
output = lokasi+dr+"\\"+m
shutil.copy(filenya, output)
wb = op.load_workbook(output)
ws = wb.active
max = ws.max_row - 1
ftext = sf()
stemmer = ftext.create_stemmer()
a = 0

for aa in ws:
    a += 1
    if a == 1:
        for bb in aa:
            u = bb.value
            p = bb.coordinate
            if u == head:
                o = 1
                for _ in range(max):
                    o += 1
                    kolom.append(f"{p[:1]}{o}")

n1 = re.compile(r"- png.*$")
n2 = re.compile(r'\b\w*([kw|wk])\w*\b', flags=re.IGNORECASE)

for k in kolom:
    norm1 = re.sub(n1, "", ws[k].value)
    norm2 = re.sub(n2, "", norm1)
    ws[k].value = stemmer.stem(norm2)

print(f"{tgl}\nProses {ket} {len(kolom)} data\n")

if os.path.exists(output):
    while True:
        if not os.path.exists(output):
            wb.save(output)
            size = os.stat(output)
            if size.st_size <= (5*1024):
                os.remove(output)
                wb.save(output)
                continue
            else:
                print(f"Silahkan cek file output {ket} pada {output}\n")
                break
        os.remove(output)
else:
    wb.save(output)
    size = os.stat(output)
    if size.st_size <= (5*1024):
        while True:
            os.remove(output)
            wb.save(output)
            print(f"Silahkan cek file output {ket} pada {output}\n")
            break
    else:
        print(f"Silahkan cek file output {ket} pada {output}\n")
```

```python
#K-fold validation
from openpyxl import Workbook
from sklearn.model_selection import cross_val_score, KFold, train_test_split
from sklearn.naive_bayes import MultinomialNB, BernoulliNB, GaussianNB

opini = []
label = []
hasil = []
opn = []
lbl = []
teks = ""
nb = "k-fold_validation"
dr = "\\sc"
lokasi = os.getcwd()
f = "output_normalisasi.xlsx"
now = dt.now()
tgl = f"\nTanggal dan waktu: {now}\n"
temp = "temp.xlsx"
ket = "K-fold validation"
filenya = lokasi+dr+"\\"+f
head1 = "opini"
head2 = "label"
m = "output_"+nb+".xlsx"
output = lokasi+dr+"\\"+m
tmp = lokasi+dr+"\\"+temp
shutil.copy(filenya, tmp)
shutil.copy(tmp, output)
vectorizer = TfidfVectorizer()
wb = op.load_workbook(tmp)
ws = wb.active
max = ws.max_row - 1
a = 0

for aa in ws:
    a += 1
    if a == 1:
        for bb in aa:
            u = bb.value
            p = bb.coordinate
            if u == head1:
                o = 1
                for _ in range(max):
                    o += 1
                    opini.append(f"{p[:1]}{o}")

a = 0
for cc in ws:
    a += 1
    if a == 1:
        for vv in cc:
            u = vv.value
            p = vv.coordinate
            if u == head2:
                o = 1
                for _ in range(max):
                    o += 1
                    label.append(f"{p[:1]}{o}")

print(f"{tgl}\nProses {ket} dari {len(opini)} data\n")

models = {
    'Multinomial': MultinomialNB(),
    'Bernoulli': BernoulliNB(),
    'Gaussian': GaussianNB()
}

for op, lb in zip(opini, label):
    opn.append(ws[op].value)
    lbl.append(ws[lb].value)

for mdlName, model in models.items():
    X_train, X_test, y_train, y_test = train_test_split(opn, lbl, test_size=0.2, random_state=42)
    xx = vectorizer.fit_transform(X_train).toarray()
    yy = y_train
    scores = cross_val_score(model, xx, yy, cv=KFold(n_splits=3))
    hasil.append({"model": mdlName, "score": scores, "average": scores.mean()})

newWB = Workbook()
sheet = newWB.active
hd = ["model", "score", "average"]
sheet.append(hd)

yy = 2
for r in hasil:
    for scr in r['score']:
        aKOL = f"A{yy}"
        bKOL = f"B{yy}"
        cKOL = f"C{yy}"
        mdl = r['model']
        avg = r['average']
        sheet[aKOL] = mdl
        sheet[bKOL] = scr
        sheet[cKOL] = avg
        yy += 1

if os.path.exists(output):
    os.remove(output)
    while True:
        if not os.path.exists(output):
            newWB.save(output)
            os.remove(tmp)
            print(f"Silahkan cek file output {ket} pada {output}\n")
            break
        else:
            os.remove(output)
            continue
else:
    wb.save(output)
    os.remove(tmp)
    print(f"Silahkan cek file output {ket} pada {output}\n")
```

```python
#Klasifikasi dengan naïve bayes
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import pandas as pd

opini = []
label = []
hasil = []
arkidf = []
ariidf = []
dr = "\\sc"
lokasi = os.getcwd()
kf = "output_k-fold_validation.xlsx"
tidf = "output_TF-IDF.xlsx"
f = "output_normalisasi.xlsx"
now = dt.now()
tgl = f"\nTanggal dan waktu: {now}\n"
ket = "klasifikasi naive bayes"
filenya = lokasi+dr+"\\"+f
kfile = lokasi+dr+"\\"+kf
tfile = lokasi+dr+"\\"+tidf

def cekmodel(mdl):
    if mdl == 'Multinomial':
        return MultinomialNB()
    elif mdl == 'Bernoulli':
        return BernoulliNB()
    elif mdl == 'Gaussian':
        return GaussianNB()

dkf = pd.read_excel(kfile)
maxavg = dkf['average'].max()
model = dkf.loc[dkf['average'] == maxavg, 'model'].iloc[0]
print(f"{tgl}\nDari step sebelumnya didapatkan model terbaik adalah {model}, dengan average bernilai {maxavg}\n")

df = pd.read_excel(filenya)
jb = df['opini'].shape[0]

ttidf = pd.read_excel(tfile)

for tt in ttidf['kata']:
    arkidf.append(tt)
print(f"Proses konvert {jb} data berdasarkan pembobotan sebelumnya\n")

for lb in df['label']:
    label.append(lb)
    
p = 0
for nn in df['opini']:
    cek = nn.split(' ')
    opini.append([])
    for cc in cek:
        if cc in arkidf and cc != '':
            nidf = ttidf.loc[ttidf['kata'] == cc,'idf'].iloc[0]
            opini[p].append(nidf)
    p += 1

e = 0
for jj in opini:
    hasil.append(sum(opini[e]))
    e += 1
    
print(f"Proses {ket} dengan menggunakan model {model} untuk {jb} data\n")

y = df['label'].values
X = np.array(hasil)
X_train, X_test, y_train, y_test = train_test_split(X.reshape(-1, 1), y, test_size=0.2, random_state=42)
ntl = (y_test == 'Netral').sum()
pos = (y_test == 'Positif').sum()
neg = (y_test == 'Negatif').sum()
ntll = (y_train == 'Netral').sum()
poss = (y_train == 'Positif').sum()
negg = (y_train == 'Negatif').sum()
print(f"Jumlah data uji: {len(X_test)}\nJumlah data latih: {len(X_train)}\nJumlah data uji netral: {ntl}\nJumlah data uji positif: {pos}\nJumlah data uji negatif: {neg}\nJumlah data latih netral: {ntll}\nJumlah data latih positif: {poss}\nJumlah data latih negatif: {negg}")
mdel = cekmodel(model)
mdel.fit(X_train, y_train)
y_pred = mdel.predict(X_test)
report = classification_report(y_test, y_pred)
print(report)
```
